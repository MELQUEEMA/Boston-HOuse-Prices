#Annotating imagees 

```python
applo_img = cv2.imread('/content/Apollo_11_Launch.jpg',cv2.IMREAD_COLOR)
plt.imshow(applo_img);
```

```python
applo_img_copy = applo_img.copy()
```

```python
applo_img_copyCircle = applo_img.copy()
```

```python
#The AA in  line_AA stands for anti aliesed 
cv2.line(applo_img_copy,(200,100),(400,100),(0,255,255),thickness=5,lineType = cv2.LINE_AA )
plt.imshow(applo_img_copy);
applo_img_rect = applo_img.copy()
```

```python
#annotation an image with a cirlce : 
applo_img_copyCircle = applo_img.copy()
cv2.circle(applo_img_copyCircle, (600,300), 100, (0,255,255), thickness=5, lineType= cv2.LINE_AA);
plt.imshow(applo_img_copyCircle)
```

```python
#The poiunt in the rectangle refer to the top left corner and the bottom left corner:
cv2.rectangle(applo_img_rect, (500, 100), (700,600), (255, 0, 255), thickness=5, lineType=cv2.LINE_8);
plt.imshow(applo_img_rect)
```

```python
imageText = applo_img.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

# Display the image
plt.imshow(imageText[:,:,::-1])
```

# Image enhancement

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img = cv2.imread('/content/New_Zealand_Coast.jpg',cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

```python
matrix = np.ones(img_rgb.shape, dtype = "uint8")*50
bright_img = cv2.add(img_rgb,matrix)
darker_img = cv2.subtract(img_rgb,matrix)
# Show the images
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(darker_img);  plt.title("Darker");
plt.subplot(132); plt.imshow(img_rgb);     plt.title("Original");
plt.subplot(133); plt.imshow(bright_img);  plt.title("Brighter");
```

Multiplications or contrast

```python
#For a darker contrast image
matrix1 = np.ones(img_rgb.shape) * .8

#For brighter contrast image
matrix2 = np.ones(img_rgb.shape) * 1.2

#To change out the contrast of the image :
img_rgb_darker   = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_brighter = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), matrix2),0,255))

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_darker);    plt.title("Lower Contrast");
plt.subplot(132); plt.imshow(img_rgb_brighter);  plt.title("brighter Contrast");
plt.subplot(133); plt.imshow(img_rgb);           plt.title("Orginal Image");
```
